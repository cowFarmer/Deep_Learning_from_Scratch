# 1.4.2 간단한 클래스 만들기

```python
class Man:
    def __init__(self, name):
        self.name = name
        print("Init")

    def hello(self):
        print("Hello " + self.name + "!")

    def goodbye(self):
        print("Good bye " + self.name + "!")

m = Man('david')
m.hello()
m.goodbye()
```

# 1.5 넘파이

```python
import numpy as np
```

```python
x = np.array([1.0, 2.0, 3.0])
print(x)

print(type(x))
```

```python
x = np.array([1.0, 2.0, 3.0])
y = np.array([2.0, 4.0, 6.0])
print(x + y)
print(x - y)
print(x / y)
```

```python
x = np.array([1.0, 2.0, 3.0])
print(x / 2.0)
```

넘파이 배열 x와 y의 개수가 같다. 이 경우에 산술 연산은 각 원소에 대해서 행한다. 원소 수가 다르면 오류가 발생하니 원소 수 맞추기가 중요하다.

넘파이는 넘파이 배열과 스칼라 산술 연산도 가능하다. 이 경우 연산이 스칼라값이 넘파이 배열의 원소별로 한 번씩 수행된다.

# 1.5.4 넘파이의 N차원 배열

```python
A = np.array([[1,2], [3,4]])
print(A)
print(A.shape)
print(A.dtype)
```

```python
B = np.array([[3,0], [0,6]])
print(A+B)
print(A*B)
print(A*10)
```

넘파이는 배열과 마찬가지로 shape가 같다면 산술 연산도 대응하는 원소별로 계산된다. 행렬과 스칼라값의 산술 연산도 가능하다.

# 1.5.5 브로드캐스트

```python
A = np.array([[1,2], [3,4]])
B = np.array([10,20])
print(B.shape)
print(A*B)
```

1차원 B배열이 2차원 A배열에 맞춰 변형된 후 원소별 연산이 이뤄진다.

# 1.5.6 원소 접근

```python
X = np.array([[51,55], [14,19], [0,4]])
print(X)
print(X[0])
print(X[0][1])
```

```python
for row in X:
    print(row)
```

```python
X = X.flatten()
print(X)
print(X[np.array([0,2,4])])
```

```python
print(X>15)
print(X[X>15])
```

넘파이 배열에 부등호를 사용하면 bool값이 나온다.   
넘파이 배열 안에서 부등호를 사용하면 해당하는 원소를 꺼낸다.

# 1.6 matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt

# 데이터
x = np.arange(0, 6, 0.1)
y = np.sin(x)

plt.plot(x,y)
plt.show()
```

```python
y2 = np.cos(x)

plt.plot(x, y, label='sin')
plt.plot(x, y2, linestyle='--', label='cos')
plt.xlabel('x')
plt.ylabel('y')
plt.title('sin & cos')
plt.legend()
plt.show()
```
